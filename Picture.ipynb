```python
import numpy as np
import matplotlib.pyplot as plt
```

# 1. Speed up ratio
## 1.1 grid decomposition

```python
fig, axes = plt.subplots(1,1, figsize=(8,8))
processes=[1,2,4,6,8]
s_100 = [1,1.5006,1.996,2.1577,2.098]
s_200 = [1,1.973,2.1514,2.5852,2.389]
s_300 = [1,2.146,2.39,2.5936, 2.6366]
s_500 = [1,2.227,2.481, 3.05489,3.11359 ]
plt.plot(processes, s_100, color='r',marker='o',label='100x100')
plt.plot(processes, s_200, color='b',marker='o',label='200x200')
plt.plot(processes, s_300, color='k',marker='o',label = '300x300')
plt.plot(processes, s_500, color='y',marker='o',label = '500x500')
axes.set_title("Speed up ratio using grid decomposition",fontsize=14)
axes.set_xlabel(r'[Processes]',fontsize=14)
axes.set_ylabel(r'[Speed up ratio]',fontsize=14)
plt.legend()
```

## 1.2 strip decomposition

```python
fig, axes = plt.subplots(1,1, figsize=(8,8))
processes=[1,2,3,5,7,8]
s_100 = [1,1.769,2.23,3.03163,4.097,4.14]
s_200 = [1,1.973,2.588,3.5224,4.49,5.567]
s_300 = [1,1.982,2.575,3.499, 4.396,4.4603]
s_500 = [1,2.227,2.7568, 3.41297,3.933,4.0246 ]
plt.plot(processes, s_100, color='r',marker='o',label='100x100')
plt.plot(processes, s_200, color='b',marker='o',label='200x200')
plt.plot(processes, s_300, color='k',marker='o',label = '300x300')
plt.plot(processes, s_500, color='y',marker='o',label = '500x500')
axes.set_title("Speed up ratio using strip decomposition",fontsize=14)
axes.set_xlabel(r'[Processes]',fontsize=14)
axes.set_ylabel(r'[Speed up ratio]',fontsize=14)
plt.legend()
```
